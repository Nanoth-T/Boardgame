```python
import pandas as pd
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSEn5eKtgjkKQXqe7LuNBDLuPa-jURsGbJgnk9ahhyTf1CMNVs-6KQRthYb3R69lMdn9-4yBjkJ1hKQ/pub?output=csv')
```

```python
df
```

```python
valuecatelist = []
for i in df["category"]:
    if (pd.notnull(i)):
        valuecatelist.extend(i.split(','))
valuecatelist
```

```python
valuecatelist
```

```python
import matplotlib.pyplot as plt
# นับจำนวนเกมในแต่ละหมวดหมู่
category_counts = pd.Series(valuecatelist).value_counts()

# เลือกแสดงเฉพาะ 10 อันดับแรก
top_10_categories = category_counts.head(10)

# พล็อตกราฟแท่ง
plt.figure(figsize=(12, 6))
top_10_categories.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Categories by Number of Games')
plt.xlabel('Category')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. กราฟวงกลมแสดงสัดส่วนของเกมในแต่ละครอบครัว
family_counts = df['family'].value_counts()
plt.figure(figsize=(8, 8))
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightpink']
family_counts.head(5).plot.pie(autopct='%1.1f%%', colors=colors)
plt.title('Top 5 Game Families')
plt.gcf().patch.set_facecolor('white')  # เปลี่ยนพื้นหลังเป็นสีขาว
plt.show()

```

```python
# 3. กราฟแท่งแสดงคะแนนเฉลี่ยของเกมในแต่ละหมวดหมู่ 10 อันดับแรก
category_avg_rating = df[df['category'].isin(valuecatelist)].groupby('category')['average_rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
category_avg_rating.plot(kind='bar', color='lightseagreen')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()


```

```python
# 4. กราฟแท่งแสดงคะแนนเฉลี่ยของเกมในแต่ละจำนวนผู้เล่นสูงสุด
players_avg_rating = df.groupby('max_players')['average_rating'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
players_avg_rating.plot(kind='bar', color='lightcoral')
plt.title('Top 20 Average Ratings by Max Players')
plt.xlabel('Max Players')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
playtime_avg_rating = df.groupby('pu')['playing_time'].mean()
playtime_avg_rating = playtime_avg_rating / 60 # แปลงนาทีเป็นชั่วโมง
plt.figure(figsize=(10, 6))
plt.hist(playtime_avg_rating, bins=15, color='lightseagreen', edgecolor='black')
plt.title('Histogram of Average Rating by Playing Time (hours)')
plt.xlabel('playing_time')
plt.ylabel('Average Rating')
plt.show()
```

```python
playtime_avg_rating = df.groupby('users_rated')['playing_time'].mean()
playtime_avg_rating.index = playtime_avg_rating.index / 60  # แปลงนาทีเป็นชั่วโมง
plt.figure(figsize=(10, 6))
plt.hist(playtime_avg_rating, bins=15, color='lightseagreen', edgecolor='black', range=(0, 4000))
plt.title('Histogram of Average Rating by Playing Time (hours)')
plt.xlabel('playing_time')
plt.ylabel('users_rated')
plt.show()
```

```python
age_avg_rating = df.groupby('min_age')['users_rated'].mean()
age_avg_rating_sorted = age_avg_rating.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
age_avg_rating_sorted.plot(kind='bar', color='lightpink')
plt.title('Average Rating by Minimum Recommended Age (High to Low)')
plt.xlabel('Minimum Recommended Age')
plt.ylabel('users_rated')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. กราฟแผนภูมิแสดงความสัมพันธ์ระหว่างคะแนนเฉลี่ยและจำนวนผู้ให้คะแนน
plt.figure(figsize=(10, 6))
plt.scatter(df['users_rated'], df['average_rating'], alpha=0.5, color='skyblue')
plt.title('Relationship between Average Rating and Number of Users Rated')
plt.xlabel('Number of Users Rated')
plt.ylabel('Average Rating')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# สร้างรายการของ 5 ศิลปินที่มีคะแนนเฉลี่ยสูงสุด
top_5_artists = df.groupby('artist')['average_rating'].mean().nlargest(5)

# สร้างรายการของศิลปินอื่น ๆ และรวมคะแนนเฉลี่ย
other_artists = df.groupby('artist')['average_rating'].mean().nsmallest(len(df['artist'].unique()) - 5)
other_avg_rating = other_artists.mean()

# สร้างกราฟ
plt.figure(figsize=(12, 6))
bar_plot = top_5_artists.plot(kind='bar', color='lightseagreen')
plt.title('Top 5 Artists with Highest Average Rating (Other included)')
plt.xlabel('Artist')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)

# กำหนดช่วงเเกน y
plt.ylim(5, 10)

# ใส่เลขในเเต่ละเเท่ง
for index, value in enumerate(top_5_artists):
    plt.text(index, value + 0.1, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```
